```python
from stockapp import models as stkmd
from dataapp import models as dtamd
from stockapp import tasks as stktks

import pandas as pd
import numpy as np
import json
```

```python
dfstockmeta=pd.read_csv('databases/olddata/stockmeta.csv')
dfwatchlist=pd.read_csv('databases/olddata/watchlist.csv')
dfwatchlist_stocks=pd.read_csv('databases/olddata/watchlist_stocks.csv')
```

```python
# pd.merge()
print dfwatchlist_stocks.head()
dfwatchlist['Symbol']=0

for ind in dfwatchlist.index:
    print ind,len(dfwatchlist)
    stock_ids=dfwatchlist_stocks[ dfwatchlist_stocks['watchlist_id']==dfwatchlist.loc[ind,'id'] ]['stock_id'].tolist()
    stocks=dfstockmeta[ dfstockmeta['id'].isin(stock_ids) ]['Symbol'].tolist()
    dfwatchlist.loc[ind,'Symbol']=json.dumps(stocks)
```

```python
for ind in dfstockmeta.index:
    print ind,len(dfstockmeta)
    
    obj=stkmd.Stockmeta(Company=dfstockmeta.loc[ind,'Company'],Marketcap=dfstockmeta.loc[ind,'Marketcap'],
                   Competitors=dfstockmeta.loc[ind,'Competitors'],Symbol=dfstockmeta.loc[ind,'Symbol'],
                   Sector=dfstockmeta.loc[ind,'Sector'],Industry=dfstockmeta.loc[ind,'Industry'],
                   Status=dfstockmeta.loc[ind,'status'])
    obj.save()
```

```python
stkmd.Watchlist.objects.all().delete()
for ind in dfwatchlist.index:
    print ind,len(dfwatchlist)
    
    obj=stkmd.Watchlist(Watchlist_name=dfwatchlist.loc[ind,'watchlistname'],
                        Watchlist_description=dfwatchlist.loc[ind,'watchlist_descrip'])
    obj.save()
    symbols=json.loads( dfwatchlist.loc[ind,'Symbol'] )

    stocks=stkmd.Stockmeta.objects.filter( Symbol__in= symbols )
    obj.Symbol.add(*stocks)
    obj.save()
    
# class Watchlist(hmd.UserBase):
# 	Watchlist_name=models.CharField(max_length=50,null=True)
# 	Watchlist_description=models.CharField(max_length=1000,null=True,blank=True)
# 	Symbol = models.ManyToManyField(Stockmeta)

# 	created_at = models.DateTimeField(auto_now_add=True,null=True)
# 	updated_at = models.DateTimeField(auto_now=True,null=True)

	
# 	def __str__(self):
# 		return self.Watchlist_name
```
